```python
import math
import numpy as np
import matplotlib.pyplot as plt
import os

# --- Your fixed parameters ---
TMAX_values = [1, 3, 9]
BarrierLength = 500
BeachWidth = 6
BermEl = 1.9
SL = 0.36
output_dir = '/Users/rsahrae/PycharmProjects/CASCADE_Coupling/CASCADE/data/Inlet_dataset'

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# --- Main Looping Logic ---

# Loop through each of your 4 barrier segments
for iB3D in range(cascade_model._ny): # <<< CHANGE HERE
    print(f"--- Processing Segment {iB3D} ---")

    # Access the data for the current segment in the loop
    DomainTS = cascade_model.barrier3d[iB3D].DomainTS   # <<< CHANGE HERE
    DuneDomain = cascade_model.barrier3d[iB3D].DuneDomain # <<< CHANGE HERE
    x_s_TS = cascade_model.barrier3d[iB3D].x_s_TS       # <<< CHANGE HERE

    # Loop through each TMAX value for the current segment
    for t in TMAX_values:
        # Create beach elevation profile
        BeachDomain = np.zeros([BeachWidth, BarrierLength // 10])
        berm = math.ceil(BeachWidth * 0.65)
        BeachDomain[berm:BeachWidth, :] = BermEl
        add = (BermEl - SL) / berm
        for i in range(berm):
            BeachDomain[i, :] = SL + add * i

        # Build the elevation map
        Domain = DomainTS[t] * 10
        Dunes = (DuneDomain[t, :, :] * 10 + BermEl)
        Dunes = np.rot90(Dunes)
        Dunes = np.flipud(Dunes)
        Beach = BeachDomain
        ElevationMap = np.vstack([Beach, Dunes, Domain])
        ElevationMap[ElevationMap < 0] = -1

        # Save the elevation array with a unique name
        # <<< CHANGE HERE: Added segment index to filename
        filename = f"Elevarray_segment_{iB3D}_time_{t}.npy"
        filepath = os.path.join(output_dir, filename)
        np.save(filepath, ElevationMap)
        print(f"Saved {filepath}")

        # Create final plot for this TMAX and segment
        Domain_final = np.vstack([Beach, Dunes, Domain])

        plt.figure(figsize=(4, 6), dpi=300)
        plt.imshow(Domain_final, origin="lower", cmap="terrain", vmin=-1.1, vmax=4.0)
        plt.colorbar(label="Elevation (m)")
        plt.xlabel("Alongshore Distance (dam)")
        plt.ylabel("Cross-Shore Distance (dam)")
        plt.title(f"Segment {iB3D}, Time = {t} yrs")

        # Save the plot with a unique name
        # <<< CHANGE HERE: Added segment index to filename
        plot_filename = os.path.join(output_dir, f"Elevation_segment_{iB3D}_time_{t}.png")
        plt.savefig(plot_filename)
        plt.close() # Close the figure to free memory
        print(f"Saved plot {plot_filename}")

print("\nProcessing complete for all segments and TMAX values.")
```
