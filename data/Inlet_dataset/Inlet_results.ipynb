```python
import numpy as np
import matplotlib.pyplot as plt
import os

# --- 1. Define the path to your output file ---
output_dir = '/Users/rsahrae/PycharmProjects/CASCADE_Coupling/CASCADE/data/Inlet_dataset/Run_Output'
file_name = '/CASCADE/data/Inlet_dataset/Run_Output/Connected_PeaIsland_5_Domain_Natural_Run.npz'  # 👈 *** REPLACE WITH YOUR FILENAME IF DIFFERENT ***
full_path = os.path.join(output_dir, file_name)

# --- 2. Load the .npz file and extract the cascade object ---
print(f"📂 Loading data from {full_path}...")
data = np.load(full_path, allow_pickle=True)
cascade_obj = data['cascade'][0]
print("✅ Data loaded successfully.")

# --- 3. Extract and organize BARRIER WIDTH data from ALL domains ---
all_widths = []
width_ts_m = [value * 10 for value in cascade_obj.barrier3d[1].InteriorWidth_AvgTS]
all_widths.append(width_ts_m)

# Convert the list of lists into a 2D NumPy array
# Shape will be (number_of_domains, number_of_time_steps)
width_data = np.array(all_widths)

# --- 4. Create the Space-Time Plot ---
print("📈 Generating domain plot of barrier width...")

# Create arrays for the x and y axes
time_axis_yr = np.arange(cascade_obj.time_step_count)
domain_spacing_m = cascade_obj.brie._dy
y_axis_m = np.arange(len(cascade_obj.barrier3d)) * domain_spacing_m

# Create the plot
fig, ax = plt.subplots(figsize=(10, 5))
# Use pcolormesh to plot the 2D width data
im = ax.pcolormesh(time_axis_yr, y_axis_m, width_data, shading='auto', cmap='magma') # Using a different colormap

# Add a color bar to show what the colors mean
cbar = fig.colorbar(im, ax=ax)
cbar.set_label('Average Barrier Width (m)') # ❗️ CHANGE: Updated label

# Label axes and add a title
ax.set_xlabel('Time (years)')
ax.set_ylabel('Alongshore Distance (m)')
ax.set_title('Barrier Width Evolution Over Time (Space-Time Plot)') # ❗️ CHANGE: Updated title

# --- 5. Save the plot ---
save_path = os.path.join(output_dir, "Domain_Width_Evolution_Plot.png")
plt.savefig(save_path, dpi=300, bbox_inches='tight')
plt.show()

print(f"🎉 Plot saved to {save_path}")
```
