```python
from noaa_coops import Station 
from datetime import datetime, timedelta
import matplotlib.pyplot as plt 
import numpy as np 
import csv 
import re
import pandas as pd 
# from fuzzywuzzy import process
 
```

```python
Hatteras = Station (id="8654467") # Set tide gauge ID
start_date = '20240101' # August 03, 2011
end_date = '20241231' # August 21, 2011

measured_water_lvls_hatteras_NAVD = Hatteras.get_data(begin_date=start_date,
                                        end_date=end_date,
                                        product="water_level",
                                        datum = 'NAVD',
                                        units="metric",
                                        time_zone= 'gmt')
measured_water_lvls_hatteras_NAVD.index = measured_water_lvls_hatteras_NAVD.index.tz_localize('UTC' )
```

```python
Hatteras = Station (id="8654467") # Set tide gauge ID
start_date = '20240101' # August 03, 2011
end_date = '20241231' # August 21, 2011

measured_water_lvls_hatteras_MHW = Hatteras.get_data(begin_date=start_date,
                                        end_date=end_date,
                                        product="water_level",
                                        datum = 'MHW',
                                        units="metric",
                                        time_zone= 'gmt')
measured_water_lvls_hatteras_MHW.index = measured_water_lvls_hatteras_MHW.index.tz_localize('UTC' )
```

```python

#plots
plt.plot(measured_water_lvls_hatteras_NAVD.index, measured_water_lvls_hatteras_NAVD['v'], label= 'Hatteras- WL NAVD')
plt.plot(measured_water_lvls_hatteras_MHW.index, measured_water_lvls_hatteras_MHW['v'], label= 'Hatteras- WL-MHW')

plt.xlabel('Dates')  # Rotates the label vertically
plt.ylabel('Measured Water Level')
plt.title('Water Levels Measured by NOAA Station 8656483')
plt.grid()
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
